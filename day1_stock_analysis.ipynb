```python
import yfinance as yf
import pandas as pd
import plotly.express as px
```

```python
hisseler = ["AAPL", "TSLA", "MSFT", "GOOGL", "AMZN"]

veri = yf.download(hisseler, start="2022-01-01", end="2024-12-31")["Close"]

print(veri.head())
```

```python
getiri = veri.pct_change().dropna()
print(getiri.describe())
```

```python
normalize = (veri / veri.iloc[0]) * 100
fig = px.line(
    normalize,
    title="5 Hisse Senedi Performans Karşılaştırması (2022-2024)",
    labels={"value": "Normalize Fiyat (Başlangıç=100)", "Date": "Tarih"}
)
fig.show()
```

```python
aapl = veri["AAPL"]

aapl_ma20 = aapl.rolling(window=20).mean()

import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=aapl.index, y=aapl, name="AAPL Fiyat"))
fig.add_trace(go.Scatter(x=aapl_ma20.index, y=aapl_ma20, name="20 Günlük Ortalama"))
fig.update_layout(title="AAPL Fiyat ve Hareketli Ortalama")
fig.show()
```
