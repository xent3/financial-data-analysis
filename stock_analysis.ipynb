```python
import yfinance as yf
import pandas as pd
import plotly.express as px
```

```python
hisseler = ["AAPL", "TSLA", "MSFT", "GOOGL", "AMZN"]

veri = yf.download(hisseler, start="2022-01-01", end="2024-12-31")["Close"]

print(veri.head())
```

```python
getiri = veri.pct_change().dropna()
print(getiri.describe())
```

```python
normalize = (veri / veri.iloc[0]) * 100
fig = px.line(
    normalize,
    title="5 Hisse Senedi Performans Karşılaştırması (2022-2024)",
    labels={"value": "Normalize Fiyat (Başlangıç=100)", "Date": "Tarih"}
)
fig.show()
```

```python
aapl = veri["AAPL"]

aapl_ma20 = aapl.rolling(window=20).mean()

import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=aapl.index, y=aapl, name="AAPL Fiyat"))
fig.add_trace(go.Scatter(x=aapl_ma20.index, y=aapl_ma20, name="20 Günlük Ortalama"))
fig.update_layout(title="AAPL Fiyat ve Hareketli Ortalama")
fig.show()
```

```python
ortalama_getiriler = getiri.mean()
print(ortalama_getiriler)
print("En yüksek ortalama getiri:", ortalama_getiriler.idxmax())
```

```python
def kategori_belirle(deger):
    if deger > 0.001:
        return "Pozitif gün"
    elif deger < -0.001:
        return "Negatif gün"
    else:
        return "Nötr gün"

appl = getiri["AAPL"].apply(kategori_belirle)
amzn = getiri["AMZN"].apply(kategori_belirle)
googl = getiri["GOOGL"].apply(kategori_belirle)
msft = getiri["MSFT"].apply(kategori_belirle)
tsla = getiri["TSLA"].apply(kategori_belirle)

print(appl.value_counts())
print(amzn.value_counts())
print(googl.value_counts())
print(msft.value_counts())
print(tsla.value_counts())
```

```python
kategori_df = getiri.apply(lambda kolon: kolon.apply(kategori_belirle))

print(kategori_df.head())
```

```python
pozitif_sayilar = (kategori_df == "Pozitif gün").sum()
print(pozitif_sayilar)
print("En çok pozitif gün yaşayan:", pozitif_sayilar.idxmax())
```
